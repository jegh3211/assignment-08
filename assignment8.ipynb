##새로운 키를 리스트의 맨 앞에 삽입하면 바로 연결만 바꾸면 되므로 O(1) 시간이 걸린다.하지만 맨 뒤에 삽입하려면 리스트의 끝까지 이동해야 하므로 O(n) 시간이 걸릴 수 있다.따라서 데이터가 많아질수록 삽입 시간이 더 오래 걸린다.


##체이닝에서 새로운 키를 맨 뒤에 삽입하면 리스트 끝까지 탐색해야 하므로 삽입 효율이 감소한다.  즉, 맨 앞 삽입은 O(1)이지만 맨 뒤 삽입은 O(n)이 될 수 있다.
